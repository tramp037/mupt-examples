```python
from mupt.reactions.polymerizeit.make_pi import make_pi

inputs = make_pi(inputs_file="inputs.yaml")
```

```python
# draw inputs['mon_A_mol'], inputs['mon_B_mol'], and inputs['dim_mol'] using rdkit and display them in the notebook
smiles_list = [inputs['mon_A_smi'], inputs['mon_B_smi'], inputs['dim_smi']]
rdkit_mols = [rdkit.Chem.MolFromSmiles(smiles) for smiles in smiles_list]

import rdkit

for mol in rdkit_mols:
    display(rdkit.Chem.Draw.MolToImage(mol))
```
